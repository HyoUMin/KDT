<h1>데이터 전처리를 위한 자료형 변환</h1> 

```python
import pandas as pd
import numpy as np
```

DF / Series의 원소의 실제 데이터에 맞게 타입 변환  
현재 타입 확인 속성: dtypes/dtypes  
현재 타입 변경 메소드: 변수명.astype(변환타입명)  



```python
#  데이터 준비 및 DF 인스턴스 생성
df = pd.DataFrame({
    "이름": ["김민수", "이서연", "박지훈"],
    "나이": ["20", "21", np.nan],
    "점수": ["85.5", "90.0", "78.5"],
    "가입일": ["2024-01-10", "2024-02-15", "2024-03-01"]
})
```

DF/Series.astype(변환타입)

```python
# # 실제 데이터와 타입이 안맞을 경우 
# # EX) 나이, 점수, 가입일 -> 결측치 있으면 오류
# df["나이"] = df["나이"].astype(int)
# df["점수"] = df["점수"].astype(float)
# df["가입일"] = df["가입일"].astype("datetime64[ns]") 
```

```python
# 실제 데이터와 타입이 안맞을 경우
# EX) 나이, 점수, 가입일
df["나이"] = df["나이"].astype('Int32')
df["점수"] = df["점수"].astype('Float32')
df["가입일"] = df["가입일"].astype("datetime64[ns]") # 권장하지 않음.
```

```python
df
```

```python
df.dtypes
```

숫자형으로 변환 함수 >>> pandas.to_numeric(데이터)

```python
#  데이터 준비 및 DF 인스턴스 생성
df = pd.DataFrame({
    "이름": ["김민수", "이서연", "박지훈"],
    "나이": ["20", "21", np.nan],
    "점수": ["85.5", "90.0", "78.5"],
    "가입일": ["2024-01-10", "2024-02-15", "2024-03-01"]
})
```

```python
df['점수']= pd.to_numeric(df['점수'], downcast='float')
```

```python
df.dtypes
```

날짜/시간 데이터 형변환 >>> pandas.to_datetime(데이터)

```python
# 가입일 데이터 형변환
df['가입일'] = pd.to_datetime(df['가입일'], dayfirst=True, format='mixed')
```

```python
df.dtypes
```
