```python
#  데이터 준비 및 DF 인스턴스 생성
DATA_FILE = '../DATA/iris.csv'

# 컬럼 인덱스 있음. 구분자는 콤마임.
iris_df = pd.read_csv(DATA_FILE)
```

variety 컬럼 object >>> 숫자/정수 라벨 인코딩

```python
# 품종명에 설정될 값 매핑
irisDF_map = {
"Setosa": 0,
"Versicolor": 1,
"Virginica": 2
}

## 결측값 NaN에는 함수 적용하여 NaN. 결과를 새로운 컬럼으로 추가
iris_df["variety_label"] = iris_df['variety'].map(irisDF_map)
## 결측값 NaN에는 함수를 적용하지 않음
iris_df["variety_label"] = iris_df['variety'].map(irisDF_map, na_action='ignore')
```

특정 값만 1로 설정하는 인코딩: 원-핫 인코딩

```python
iris_df2 = pd.get_dummies(iris_df, columns=["variety"], dtype=int)
```
