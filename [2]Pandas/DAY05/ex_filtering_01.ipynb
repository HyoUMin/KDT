<h1>데이터 전처리를 위한 필터링</h1>  

DF/Series에서 원하는 데이터만 추출  
불린 인덱싱
isin()

```python
import pandas as pd
import numpy as np


DATA_FILE = '../DATA/학생관리부.xlsx'

df = pd.read_excel(DATA_FILE, header = 2)
```

```python
df
```

```python
# 특정 조건에 맞는 데이터만 추출. 즉, 필터링.
# 석차 상위 10위까지만 데이터 추출
# 비교 연산으로 T/F로 구성된 Series 완성
b = df[df.columns[-1]] <= 10
b
```

```python
# 특정 조건에 맞는 데이터만 추출. 즉, 필터링.
# 석차 상위 10위까지만 데이터 추출
# 비교 연산으로 T/F로 구성된 Series 완성
b = df[df.columns[-1]] <= 10
df[b]
```

```python
# 현재 결과를 반대로
df[~b]
```

```python
df.columns[1], df.columns[-1]
```

```python
# 학년 컬럼
df[df.columns[1]] == 1
```

```python
# 석차 컬럼 <= 10
df[df.columns[-1]] <= 10
```

```python
# 1학년 중에 상위 10위 안에 들어가는 학생 데이터
# 학년 컬럼 == 1, 석차  <= 10
m = (df[df.columns[1]] == 1) & (df[df.columns[-1]] <= 10)
```

```python
df[m]
```
